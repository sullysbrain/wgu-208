### Cleaning Missing Attribute: Internet Service

```python
# Internet Service Attribute Analysis - Exploration

df_clean['InternetService'].info()

print(df_clean['InternetService'])
print(df_clean['InternetService'].value_counts())
```

```python
# Internet Service Cleaning - 
# Calculate Percentages of Values and Fill Nulls with Same

# Calculate distribution of known values
dist = df_clean['InternetService'].value_counts(normalize=True)

# Count Nulls to impute based on the distribution
null_count = df_clean['InternetService'].isnull().sum()
impute_counts = (dist * null_count).round().astype(int)

# Impute missing values based on the distribution
missing_indices = df['InternetService'].isnull()
df_clean.loc[missing_indices, 'InternetService'] = np.random.choice(dist.index, size=null_count, p=dist.values)

# Verify the imputation
print(df_clean['InternetService'].value_counts(normalize=True))
df_clean['InternetService']
```

## E. Principal Component Analysis

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import seaborn as sns
```

```python
df_clean.info()
```

### Select Attributes and Normalize

```python
column_names = df_clean.columns
dtypes = df_clean.dtypes
first_row = df_clean.iloc[1].values
for column_name, dtype, value in zip(column_names, dtypes, first_row):
    print(f"{column_name:<15}\t{dtype}\t{value:<18}")    
```

```python
# Selection of Attributes for PCA to only include Quantitative Attributes
df_pca = df_clean[['Population', 'Children', 'Age', 'Income', 'Outage_sec_perweek', 'Email',
                  'Yearly_equip_failure', 'Tenure', 'MonthlyCharge', 'Bandwidth_GB_Year']]
df_pca_normalized = (df_pca - df_pca.mean()) / df_pca.std() 
df_pca_normalized.head(5)
```

```python
# Create PCA
pca = PCA(n_components = df_pca.shape[1])
pca.fit(df_pca_normalized)
```

```python
# Convert PCA to dataframe
pca_df = pd.DataFrame(pca.transform(df_pca_normalized), 
                      columns=['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6', 
                                'PC7', 'PC8', 'PC9', 'PC10'])
```

### E1. Loading Output

```python
loadings = pd.DataFrame(pca.components_.T,
                        columns = ['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6', 
                                   'PC7', 'PC8', 'PC9', 'PC10'],
                        index = df_pca_normalized.columns)
loadings
```

### Kaiser Rule Applied to PCs (with eigenvalues creation)

```python
import numpy as np
cov_matrix = np.dot(df_pca_normalized.T, df_pca_normalized) / pca_df.shape[0]
eigenvalues = [np.dot(eigenvector.T, np.dot(cov_matrix, eigenvector)) for eigenvector in pca.components_]

plt.plot(eigenvalues)
plt.title('Eigenvalues for PCA')
plt.xlabel('# of components')
plt.ylabel('eignenvalues')
plt.axhline(y=1, color='red')
plt.show()
```
